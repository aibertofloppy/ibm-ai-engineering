## Ejercicio 2: Función con parámetro

Crea una función `imprimir_dos_veces(texto)` que reciba un string y lo imprima dos veces.

```python
# Tu solución aquí
def imprimir_dos_veces(texto):
    print(texto)
    print(texto)

imprimir_dos_veces("Hola")
```

## Ejercicio 3: Función que combina otras funciones

Crea una función `linea()` que imprima 40 guiones y úsala en otra función llamada `separador()` que imprima una línea y luego un mensaje.

```python
# Tu solución aquí
def linea():
    print("-" * 40)

def separador():
    linea()
    print("Fin del bloque")
    linea()

separador()
```

## Ejercicio 4: Generalizar funciones

Modifica la función `linea()` para que reciba un **carácter** y un **número** que indique cuántas veces repetirlo.

```python
# Tu solución aquí
def linea(caracter, cantidad):
    print(caracter * cantidad)

linea("*", 30)
linea("=", 50)
```
